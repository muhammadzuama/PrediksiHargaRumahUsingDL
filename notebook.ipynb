Data process

```python
# Tentukan fitur (features) dan label (target)
features = df_new.columns.tolist()  # Ambil semua nama kolom
features.remove("price")            # Hapus kolom target dari daftar fitur
target = df_new["price"]            # Tentukan target sebagai series

# Cek hasil
print("Features:", features)  # Menampilkan daftar fitur
print("Target:", target.name) # Menampilkan nama target
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Model Deep Learning
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Bidirectional, GRU, Dense, Dropout

# Mengabaikan warning TensorFlow agar output lebih rapi
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.preprocessing import MinMaxScaler

# Ambil hanya fitur numerik dari df_new
X = df_new[features]  # Menggunakan nama kolom yang sudah dipilih sebagai fitur
y = df_new["price"].values.reshape(-1, 1)  # Ubah target menjadi array 2D

# Normalisasi data
scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)

scaler_y = MinMaxScaler()
y_scaled = scaler_y.fit_transform(y)

# Cek hasil
print("X_scaled shape:", X_scaled.shape)  # Bentuk array fitur setelah normalisasi
print("y_scaled shape:", y_scaled.shape)  # Bentuk array target setelah normalisasi
```

```python

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# Ubah bentuk data agar sesuai untuk model RNN (samples, timesteps, features)
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))
```

```python
from tensorflow.keras import backend as K
from sklearn.metrics import r2_score

# Fungsi custom untuk menghitung R2
def r2_metric(y_true, y_pred):
    return K.variable(r2_score(y_true, y_pred))

# # Kompilasi model dengan metrik mae, mse, dan r2
```

```python
def build_lstm():
    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=(1, len(features))),
        LSTM(32, return_sequences=False),
        Dense(16, activation='relu'),
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

lstm_model = build_lstm()
```

```python
def build_bilstm():
    model = Sequential([
        Bidirectional(LSTM(64, return_sequences=True), input_shape=(1, len(features))),
        Bidirectional(LSTM(32, return_sequences=False)),
        Dense(16, activation='relu'),
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

bilstm_model = build_bilstm()
```

```python
def build_gru():
    model = Sequential([
        GRU(64, return_sequences=True, input_shape=(1, len(features))),
        GRU(32, return_sequences=False),
        Dense(16, activation='relu'),
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

gru_model = build_gru()
```

```python
def build_bigru():
    model = Sequential([
        Bidirectional(GRU(64, return_sequences=True), input_shape=(1, len(features))),  # BiGRU pertama
        Bidirectional(GRU(32, return_sequences=False)),  # BiGRU kedua
        Dense(16, activation='relu'),  # Fully connected layer
        Dropout(0.2),  # Regularization with Dropout
        Dense(1)  # Output layer (untuk prediksi harga rumah)
    ])
    
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])  # Compile the model with Adam optimizer
    return model

bigru_model = build_bigru()
```

```python
history_lstm = lstm_model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test), verbose=1)
history_bilstm = bilstm_model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test), verbose=1)
history_gru = gru_model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test), verbose=1)
history_bigru = bigru_model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test), verbose=1)
```

```python
import tensorflow as tf
from tensorflow.keras import backend as K

# Corrected custom R2 metric function using TensorFlow operations
def r2_metric(y_true, y_pred):
    # Calculate the residuals
    residual = y_true - y_pred
    # Calculate the total sum of squares
    ss_total = K.sum(K.square(y_true - K.mean(y_true)))
    # Calculate the residual sum of squares
    ss_residual = K.sum(K.square(residual))
    # Calculate RÂ²
    return 1 - ss_residual / (ss_total + K.epsilon())

# Compile the models with mse, mae, and the corrected r2 metric
lstm_model.compile(optimizer='adam', loss='mse', metrics=['mae', 'mse', r2_metric])
bilstm_model.compile(optimizer='adam', loss='mse', metrics=['mae', 'mse', r2_metric])
gru_model.compile(optimizer='adam', loss='mse', metrics=['mae', 'mse', r2_metric])
bigru_model.compile(optimizer='adam', loss='mse', metrics=['mae', 'mse', r2_metric])

# Evaluasi Model
lstm_eval = lstm_model.evaluate(X_test, y_test)
bilstm_eval = bilstm_model.evaluate(X_test, y_test)
gru_eval = gru_model.evaluate(X_test, y_test)
bigru_eval = bigru_model.evaluate(X_test, y_test)

# Menampilkan Hasil
print(f"LSTM -> Loss: {lstm_eval[0]}, MAE: {lstm_eval[1]}, MSE: {lstm_eval[2]}, R2: {lstm_eval[3]}")
print(f"BiLSTM -> Loss: {bilstm_eval[0]}, MAE: {bilstm_eval[1]}, MSE: {bilstm_eval[2]}, R2: {bilstm_eval[3]}")
print(f"GRU -> Loss: {gru_eval[0]}, MAE: {gru_eval[1]}, MSE: {gru_eval[2]}, R2: {gru_eval[3]}")
print(f"BiGRU -> Loss: {bigru_eval[0]}, MAE: {bigru_eval[1]}, MSE: {bigru_eval[2]}, R2: {bigru_eval[3]}")
```

```python
import matplotlib.pyplot as plt

# Evaluasi Model
lstm_eval = lstm_model.evaluate(X_test, y_test)
bilstm_eval = bilstm_model.evaluate(X_test, y_test)
gru_eval = gru_model.evaluate(X_test, y_test)
bigru_eval = bigru_model.evaluate(X_test, y_test)

# Menampilkan Hasil
models = ['LSTM', 'BiLSTM', 'GRU', 'BiGRU']
losses = [lstm_eval[0], bilstm_eval[0], gru_eval[0], bigru_eval[0]]
mae = [lstm_eval[1], bilstm_eval[1], gru_eval[1], bigru_eval[1]]
mse = [lstm_eval[2], bilstm_eval[2], gru_eval[2], bigru_eval[2]]
r2 = [lstm_eval[3], bilstm_eval[3], gru_eval[3], bigru_eval[3]]

# Grafik
fig, ax = plt.subplots(2, 2, figsize=(12, 10))

# Loss
ax[0, 0].bar(models, losses, color='skyblue')
ax[0, 0].set_title('Loss Comparison')
ax[0, 0].set_ylabel('Loss')

# MAE
ax[0, 1].bar(models, mae, color='salmon')
ax[0, 1].set_title('MAE Comparison')
ax[0, 1].set_ylabel('MAE')

# MSE
ax[1, 0].bar(models, mse, color='lightgreen')
ax[1, 0].set_title('MSE Comparison')
ax[1, 0].set_ylabel('MSE')

# R2
ax[1, 1].bar(models, r2, color='lightcoral')
ax[1, 1].set_title('R2 Comparison')
ax[1, 1].set_ylabel('R2')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Tentukan jumlah fold
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Simpan hasil evaluasi
lstm_scores = []
bilstm_scores = []
gru_scores = []
bigru_scores = []

lstm_mse = []
bilstm_mse = []
gru_mse = []
bigru_mse = []

lstm_r2 = []
bilstm_r2 = []
gru_r2 = []
bigru_r2 = []

# Looping untuk setiap fold
for train_index, val_index in kf.split(X_scaled):
    X_train, X_val = X_scaled[train_index], X_scaled[val_index]
    y_train, y_val = y_scaled[train_index], y_scaled[val_index]
    
    # Ubah bentuk data agar sesuai dengan input model RNN
    X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
    X_val = X_val.reshape((X_val.shape[0], 1, X_val.shape[1]))
    
    # LSTM
    lstm_model = build_lstm()
    lstm_model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=0, validation_data=(X_val, y_val))
    y_pred = lstm_model.predict(X_val)
    lstm_score = lstm_model.evaluate(X_val, y_val, verbose=0)
    lstm_scores.append(lstm_score[1])  # Simpan MAE
    lstm_mse.append(mean_squared_error(y_val, y_pred))
    lstm_r2.append(r2_score(y_val, y_pred))
    
    # BiLSTM
    bilstm_model = build_bilstm()
    bilstm_model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=0, validation_data=(X_val, y_val))
    y_pred = bilstm_model.predict(X_val)
    bilstm_score = bilstm_model.evaluate(X_val, y_val, verbose=0)
    bilstm_scores.append(bilstm_score[1])
    bilstm_mse.append(mean_squared_error(y_val, y_pred))
    bilstm_r2.append(r2_score(y_val, y_pred))
    
    # GRU
    gru_model = build_gru()
    gru_model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=0, validation_data=(X_val, y_val))
    y_pred = gru_model.predict(X_val)
    gru_score = gru_model.evaluate(X_val, y_val, verbose=0)
    gru_scores.append(gru_score[1])
    gru_mse.append(mean_squared_error(y_val, y_pred))
    gru_r2.append(r2_score(y_val, y_pred))
    
    # BiGRU
    bigru_model = build_bigru()
    bigru_model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=0, validation_data=(X_val, y_val))
    y_pred = bigru_model.predict(X_val)
    bigru_score = bigru_model.evaluate(X_val, y_val, verbose=0)
    bigru_scores.append(bigru_score[1])
    bigru_mse.append(mean_squared_error(y_val, y_pred))
    bigru_r2.append(r2_score(y_val, y_pred))

# Hitung rata-rata MAE, MSE, dan R2 dari setiap model
print("LSTM MAE (10-Fold CV):", np.mean(lstm_scores))
print("LSTM MSE (10-Fold CV):", np.mean(lstm_mse))
print("LSTM R2 (10-Fold CV):", np.mean(lstm_r2))

print("BiLSTM MAE (10-Fold CV):", np.mean(bilstm_scores))
print("BiLSTM MSE (10-Fold CV):", np.mean(bilstm_mse))
print("BiLSTM R2 (10-Fold CV):", np.mean(bilstm_r2))

print("GRU MAE (10-Fold CV):", np.mean(gru_scores))
print("GRU MSE (10-Fold CV):", np.mean(gru_mse))
print("GRU R2 (10-Fold CV):", np.mean(gru_r2))

print("BiGRU MAE (10-Fold CV):", np.mean(bigru_scores))
print("BiGRU MSE (10-Fold CV):", np.mean(bigru_mse))
print("BiGRU R2 (10-Fold CV):", np.mean(bigru_r2))
```

```python
import matplotlib.pyplot as plt

# Definisikan model dan hasil evaluasi
models = ["LSTM", "BiLSTM", "GRU", "BiGRU"]
mae_scores = [np.mean(lstm_scores), np.mean(bilstm_scores), np.mean(gru_scores), np.mean(bigru_scores)]
mse_scores = [np.mean(lstm_mse), np.mean(bilstm_mse), np.mean(gru_mse), np.mean(bigru_mse)]
r2_scores = [np.mean(lstm_r2), np.mean(bilstm_r2), np.mean(gru_r2), np.mean(bigru_r2)]

# Membuat plot
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].bar(models, mae_scores, color=['blue', 'green', 'red', 'purple'])
ax[0].set_title("Mean Absolute Error (MAE)")
ax[0].set_ylabel("MAE")

ax[1].bar(models, mse_scores, color=['blue', 'green', 'red', 'purple'])
ax[1].set_title("Mean Squared Error (MSE)")
ax[1].set_ylabel("MSE")

ax[2].bar(models, r2_scores, color=['blue', 'green', 'red', 'purple'])
ax[2].set_title("R-Squared (RÂ²)")
ax[2].set_ylabel("RÂ²")

plt.tight_layout()
plt.show()
```
